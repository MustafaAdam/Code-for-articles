```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns;
sns.set(palette='deep')


plt.rcParams['figure.figsize'] = 14, 12
# plt.rcParams.update({'figure.figsize': (18, 12)})
plt.rcParams.update({'font.size': 12})


np.random.seed(0)
# plt.rcParams.keys()
```

# Bias

Bias is a word in statistics and machine learning that measures how different our algorithm's predictions is from the correct outputs.<br>
In machine learning, our task is to infer the function (that can never be known explicitly) that has generated the data we have at hand.<br>
We approximate this function with a model and then see what the model produces. The difference between the model prediction and the actual true outputs of the function that generated this data is the bias.

```python
x_axis = np.linspace(0, 10, 10)
slope = 3
random_delta = np.random.normal(0, 3, size=10)
y_axis = slope * x_axis + random_delta
```

```python
plt.scatter(x_axis, y_axis, label='True output')
plt.plot(x_axis, 3 * x_axis, 'k', label='Predicted output')

plt.xlabel('Input'); plt.ylabel('Output')
plt.xticks(np.arange(0, 10, 1))
plt.title('Visual representation of bias', fontsize=16)
plt.legend();
```

# Underfitting the data

When our chosen statistical model is too simple for the data we have at hand, our model will naturally not be able to learn from the data, predict well on future data points and infer anything meaningful.
Since the model is incapable of making the right predictions, the bias will be high (Remember the bias is a measure of the wrongness in the predictions).<br>
Bias and variance has an inverse relationship. When one goes up, the other goes down.<br>
so when we have high bias due to the simplicity of the model in the face of the complexity of the data, we also get low variance.<br>

Low variance is when our model does not change much even when the data has changed significantly.<br>
A certain amount of flexibility is required from a good model in order to learn and explain the data at hand, and predict accurately on future unseen data points.<br>

Underfitting the data refers to this siutation of high bias and low variance. It's when our model is so simple and so lacking in flexibility, that it cannot digest and learn complexity in the data<br>

A graph that illustrates underfitting is when our data is generated by a sinusoidal phenomena like sound and light waves for example.<br>
But our chosen model is a linear regression.<br>
Sine waves are not linear at all. But by our choice of model, we are assuming that it is.<br>
As a result, our model will not be able to understand our data. Therefore, we will have high bias (Since it's going to have a high amount of wrong predictions).<br>
It will also have low variance because no matter how many different sine waves we throw at it, the model will not be capable of any flexibility due to its linearity.

```python
x_axis = np.linspace(0, 15, 50)
sine = np.sin(x_axis)

# This is a perfect sine wave. Let's add some random noise to it to simulate randommness in a sinusodial phenomena.
# (There will be always be a certain amount of irreducible error and this accounts for it)
random_error = np.random.normal(0, 0.5, 50)
data = sine + random_error

figure, ax = plt.subplots(1, 1)
ax.scatter(x_axis, data, label='Data')
ax.set_title('Random data generated from a sinusodial source', fontsize=16)
ax.set_xlabel('Input')
ax.set_ylabel('Sin')
ax.legend();
```

Now that we have our random sinusodial data points, let us see how well linear model will explain the data.<br>
It won't.

```python
figure, ax = plt.subplots(1, 1)

slope = 0.01
linear = slope * x_axis 

ax.scatter(x_axis, data, label='Data generated from a random sinusoidal phenomena')
ax.plot(x_axis, linear, label='Linear model', c='black')

ax.set_title('Linear model on nonlinear data', fontsize=16)
ax.set_xlabel('Input')
ax.set_ylabel('Sin')
ax.legend();
# figure.savefig('Linear model underfitting the data')
```

```python
# Another data set from the same sinusodial function

new_random  = np.random.normal(0, 0.3, 50)
second_data = sine + new_random

fig, ax = plt.subplots(1, 1)
ax.scatter(x_axis, second_data, label='Data generated from another random sinusoidal phenomena')
ax.plot(x_axis, linear, label='linear model', c='black')
ax.set_xlabel('Input')
ax.set_ylabel('Sin')
ax.set_title('Linear model is still having high bias and low flexibility(variance) to the new data', fontsize=16)
ax.legend();
# fig.savefig('Undefitting 2')
```

So our model is still not able to show sufficient amount of flexibility and variance to any new datasets.<br>
This is an inherent problem in the model because it's a linear model while our data set are sinusodial(nonlinear).<br>
This lack of flexibility is a manifestation of low variance.<br>
Also, this linear model did not learn or fit the data well.<br>
This is why it will have high bias.<br>

#### <b>High bias</b> and <b>low variance</b> is called <b>underfitting</b>

Put it another way, high wrongness and low flexibility is underfitting

# Variance

```python
plt.figure(figsize=(14, 12))

random_delta = np.random.normal(0, 0.2, size=30)

x_axis = np.linspace(0, 6.3, 30)
true_function = np.sin(x_axis) 
data = true_function + random_delta

plt.scatter(x_axis, data, c='b', label='Data points')
plt.plot(x_axis, data, 'k', label='Model predictions')

plt.xlabel('Input')
plt.ylabel('Output')
plt.title('The model is very sensitive to small and local variation in the data', fontdict={'fontsize': 20})
plt.legend();
# plt.savefig('Overfitting.png')
```

```python
x = np.linspace(0, 3, 60)
low_variance_plot = np.random.normal(2, 0.0005, 60)
high_variance_plot = np.random.normal(2, 0.05, 60)

fig, ax = plt.subplots(1, 2)

ax[0].scatter(x, low_variance_plot)
ax[0].set_title('Low Variance')
ax[1].scatter(x, high_variance_plot)
ax[1].set_title('High Variance')
fig.savefig('Variance')
```
